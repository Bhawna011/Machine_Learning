# Matplotlib

Matplotlib Basics with few examples

```python
#  Draw a line in a diagram from position (0,0) to position (6,250)
import matplotlib.pyplot as plt
import numpy as np 

xpoints=np.array([0,6])
ypoints=np.array([0,250])
plt.plot(xpoints,ypoints)
plt.show()
```

```python
# Let's assume dictionary looks like this
data={
    "TV":[100,200,300,400,500],
    "Sales":[10,20,25,30,40]
}
```

```python
# can access the "TV" and "Sales" data like this 
tv_data=data["TV"]
sales_data=data["Sales"]
```

```python
# Matplotlib to plot the data 
import matplotlib.pyplot as plt 

plt.figure(figsize=(8,4)) # set the figure size

plt.plot(tv_data, sales_data, marker="o",linestyle="-") # create the plot 

plt.title("TV vs Sales") # add a title to the plot 

plt.xlabel("TV Budget") # Label for the x-axis
plt.ylabel("Sales") # Label for the y-axis

plt.grid(True) # Add grid lines

plt.show()
```
